```python
'''
Como as notas representam o grau de satisfação do investidor com a empresa. Para que não haja distorção quanto a média do nível de satisfação e como não há outliers, 
usaremos a média para preencher os valores faltantes. Primeiramente, havíamos pensado em atribuir uma média diferente para cada perfil de investidor, mas, como pode ser observado abaixo,
não há  grandes diferenças entre a média atribuída por perfil de investidor. Sendo assim, atribuímos a média da Nota1 para os valores faltantes da Nota1 e a média da Nota2 para os valores 
faltantes da Nota2.  
'''
df_ordenado.groupby(['Perfil']).agg({'Nota1':'mean', 'Nota2':'mean'})
```

## 7 - Manipule os dados a seguir:
- Criar coluna "Total" somando o valor de todos os ativos por investidor
- Criar coluna "Media" com a média das notas 1 e 2
- Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)
- Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)
- Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar a função map)
- Transformar os dados da coluna Media em inteiro
- Criar coluna "NPS" onde o investidor é classificado como promotor se a média das notas é maior ou igual a 80, como neutro se a média estiver entre 50 e 80, e como detrator se for abaixo de 50


```python
# Criando a coluna 'Total' (soma de todos os ativos):

df['Total'] = sum((df['Ativo01'], 
                   df['Ativo02'],
                   df['Ativo03'],
                   df['Ativo04'],
                   df['Ativo05'],
                   df['Ativo06'],
                   df['Ativo07']))
```

```python
# Criando a coluna 'Media'(média das notas 1 e 2):

df['Media'] = (df['Nota1'] + df['Nota2']) / 2
```

```python
# Criando a coluna 'Dominio'(dominio do e-mail):

df['Dominio'] = [email.split('@')[1] for email in df['Email']]
```

```python
# Mudando o nome da coluna "Salário" para "Renda", tirando o caracter $ e mudando o tipo para numérico:

df = df.rename(columns = {'Salario': 'Renda'})
df['Renda'] = df['Renda'].str.strip('$')
df['Renda'] = pd.to_numeric(df['Renda'])
```

```python
# Mudando os dados da coluna "Sexo" para Masculino e Feminino.

df['Sexo'] = df['Sexo'].map({'Masc':'Maculino', 'Fem':'Feminino'})
```

```python
# Transformando os dados da coluna 'Media' em inteiro.

df = df.astype({'Media':int})
df.dtypes
```

```python
# Criando a coluna "NPS" onde o investidor é classificado como promotor se a média das notas é maior ou igual a 80, como neutro se a média estiver entre 50 e 80, 
# e como detrator se for abaixo de 50.

import numpy as np

filtros = [
    (df.Media >= 80),
    (df.Media >= 50) & (df.Media < 80),
    (df.Media < 50)
]

valores = ['promotor', 'neutro', 'detrator']

df['NPS'] = np.select(filtros, valores)
  
```
